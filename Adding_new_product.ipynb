```python
import json

Entry=True
while(Entry):
    fd=open('prodlist.json','r')
    r=fd.read()
    fd.close()
    record=json.loads(r)

    print("Press: 0  for Showing All Data.\n")
    print("Press: 1 for Add product.\n")
    print("Press: 2 for Update\n")
    print("Press: 3 for Delete.\n")
    print("Press: 4  for Exit.\n")
    x=int(input(""))
    if(x==0):
        print("******"*15)
        print("\tProduct_id\tProduct_name\tProduct_Quantity\tPrice")
        print("******"*15)
        for i in record.keys():
            print(f"\t{i}\t\t{record[i]['product_name']}\t\t{record[i]['product_quantity']}\t\t{record[i]['product_price']}\n")
    if(x==1):
        y=True
        while(y):
            found=False
            pro_id=input("Enter Product Id: ")
            for i in record.keys():
                if(i==pro_id):
                    print(f"product Id {pro_id} is already pressent\n")
                    print("Please Enter a new Id :")
                    found=True
            if(found):
                continue;
            else:
                pro_name=input("Enter Product Name:- ")
                pro_quantity=int(input("Enter Number of Quantity of product:- "))
                pro_price=int(input("Enter price of each product:- "))

                record[pro_id]={'product_name':pro_name,'product_quantity':pro_quantity,'product_price':pro_price}

                js = json.dumps(record)
                fd = open("prodlist.json",'w')
                fd.write(js)
                fd.close()
                print("******Your Product Add Success Fully******")
                y=False
    elif(x==2):
        y=True
        while(y):
            found=False
            pro_id=input("Enter Product Id: ")
            for i in record.keys():
                if(i==pro_id):
                    pro_name=input("Enter Product Name:- ")
                    pro_quantity=int(input("Enter Number of Quantity of product:- "))
                    pro_price=int(input("Enter price of each product:- "))
                    record[pro_id]={'product_name':pro_name,'product_quantity':pro_quantity,'product_price':pro_price}
                    js = json.dumps(record)
                    fd = open("prodlist.json",'w')
                    fd.write(js)
                    fd.close()
                    print("******Your Product Update Success Fully******")
                    found=True
                    y=False
            if(found):
                continue;
            else:
                print(f"product Id {pro_id} is Not pressent\n")
                print("Please Re-enter Product Id:")
    elif(x==3):
        y=True
        while(y):
            found=False
            pro_id=input("Enter Product Id: ")
            for i in record.keys():
                if(i==pro_id):
                    found=True
                    y=False
            if(found):
                del record[pro_id]
                js = json.dumps(record)
                fd = open("prodlist.json",'w')
                fd.write(js)
                fd.close()
                print("******Delete Success Fully******")
                continue;
            else:
                print(f"product Id {pro_id} is Not pressent\n")
                print("Please Re-enter Product Id:")
    elif(x==4):
        Entry=False
    else:
        print("Please Enter a valid number!!!!!\n")
    
                
        


```
